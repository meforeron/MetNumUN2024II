# 1. Cambiar técnica de ML

Se uso Random Forest sobre Bayesian Ridge Regression

1. **Relaciones no lineales:**  
   A diferencia de **Bayesian Ridge Regression**, que solo captura relaciones lineales, **Random Forest** puede identificar patrones más complejos y no lineales.

2. **Reducción de sobreajuste:**  
   Utilizando múltiples árboles, **Random Forest** reduce el riesgo de sobreajuste al promediar las predicciones, lo que lo hace más robusto ante datos ruidosos.

3. **Importancia de características:**  
   **Random Forest** identifica automáticamente las características más relevantes para las predicciones, lo que mejora la selección de variables.

En resumen, **Random Forest** es más flexible y robusto, lo que puede llevar a un modelo más preciso en comparación con **Bayesian Ridge Regression**.

# 4. Nuevas variables


1. **OBV (On Balance Volume):**  
   Mide la acumulación o distribución de un activo, utilizando el volumen para confirmar la dirección de la tendencia.

2. **SMA (Simple Moving Average):**  
   Promedia los precios de cierre en un período (20 días en este caso) para identificar la dirección de la tendencia, actuando como soporte o resistencia.

3. **ATR (Average True Range):**  
   Mide la volatilidad del mercado, ayudando a ajustar estrategias de trading según la amplitud de las fluctuaciones del precio.

Estos indicadores fueron seleccionados para proporcionar una visión completa de la tendencia, el volumen y la volatilidad del mercado.
