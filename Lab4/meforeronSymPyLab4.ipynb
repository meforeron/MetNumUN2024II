## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x9,y = y1 + I*y9, z = z1 + I*z9, x * y * z

www.wolframalpha.com_input_i=x%3Dx1+%2B+I_x9%2Cy+%3D+y1+%2B+I_y9%2C+z+%3D+z1+%2B+I_z9%2C+x+_+y+_+z.png

```
from sympy import symbols, I, N, pi, expand
x1, x9, y1, y9, z1, z9 = symbols("x1 x5 y1 y5 z1 z5", real=True)
x = x1 + I*x9
y = y1 + I*y9
z = z1 + I*z9

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=9/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
